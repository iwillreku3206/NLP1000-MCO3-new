```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
import gc

en_bcl_tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-bcl")
en_bcl_model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-bcl")

en_bcl_tokenizer.tokenize("Who are you?")
en_bcl_inputs = en_bcl_tokenizer("What are you doing?", return_tensors="pt")
bcl_outputs = en_bcl_model.generate(**en_bcl_inputs)
bcl = en_bcl_tokenizer.decode(bcl_outputs[0], skip_special_tokens=True)
                              
del en_bcl_model
del en_bcl_tokenizer
torch.cuda.empty_cache()
gc.collect()

bcl
```

```python
from datasets import Dataset
from unsloth import FastLanguageModel
import torch

bic_iva_model, bic_iva_tokenizer = FastLanguageModel.from_pretrained(
    model_name = "trained-bic-iva",
    max_seq_length = 512,   # Context length - can be longer, but uses more memory
    load_in_4bit = True,     # 4bit uses much less memory
    load_in_8bit = False,    # A bit more accurate, uses 2x memory
    full_finetuning = False, # We have full finetuning now!
)

messages = [
    {"role" : "user", "content" : f"Translate Bicol to Ivatan: {bcl}"}
]
text = bic_iva_tokenizer.apply_chat_template(
    messages,
    tokenize = False,
    add_generation_prompt = True, # Must add for generation
    enable_thinking = False, # Disable thinking
)
```

```python
from transformers import TextIteratorStreamer

streamer = TextIteratorStreamer(bic_iva_tokenizer, skip_prompt = True, skip_special_tokens = True)
_ = bic_iva_model.generate(
    **bic_iva_tokenizer(text, return_tensors = "pt").to("cuda"),
    max_new_tokens = 256, # Increase for longer outputs!
    temperature = 0.7, top_p = 0.8, top_k = 20, # For non thinking
    streamer = streamer,
)

iva = ""
for word in streamer:
    print(word, end = "", flush = True)
    iva += word

del bic_iva_model
del bic_iva_tokenizer
gc.collect()
torch.cuda.empty_cache()
```
